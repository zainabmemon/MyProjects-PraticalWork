```python
#creating train folder in flowers
train_dir=os.path.join(path,'train')
if (not(os.path.exists(train_dir))):
    os.mkdir(train_dir)
```

```python
#moving the classes to train folder
for i in range(len(classes)):
    if(os.path.exists(os.path.join(path,classes[i]))):
        shutil.move(os.path.join(path,classes[i]),train_dir)
        
```

```python
#ploting the graph of imbalanced dataset
Listx=[]
for i in range(len(classes)):
    cur_class=os.listdir(os.path.join(train_dir,classes[i]))
    print(classes[i]+" "+str(len(cur_class)))
    Listx.append(len(cur_class))
plt.figure(figsize=(5, 5)),
plt.xticks(np.arange(len(classes)), classes, rotation='horizontal')
plt.bar(classes, Listx, align='center',color=('b', 'c', 'g', 'r','y') )
plt.ylabel('No of Images')
plt.xlabel('Classes')
plt.margins(0.001)
plt.title('Imbalanced Dataset')
#plt.figure(figsize=(100,10000))
plt.show()
```

```python
#creating test folder
test_dir=os.path.join(path,'test')
if (not(os.path.exists(test_dir))):
    os.mkdir(test_dir)
```

```python
#creating each class folder in test folder 
for i in classes:
    train_classes=os.path.join(train_dir,i)
    test_classes=os.path.join(test_dir,i)
    if(not(os.path.exists(test_classes))):
        os.mkdir(test_classes)
```

```python
import random
rand=random.choice([1,2,3])
print(rand)
```

```python
#moving 20 percent of each class train images to test
if(len(os.listdir(os.path.join(train_dir,classes[0])))>760):
    for i in range(len(classes)):
        train_class_images=os.listdir(os.path.join(train_dir,classes[i]))
        print(classes[i])
        for j in range((len(train_class_images)*20)//100):
            rand=random.choice(os.listdir(os.path.join(train_dir,classes[i])))
            print(rand)
            shutil.move(os.path.join(train_dir,classes[i])+"\\"+rand,os.path.join(test_dir,classes[i]))
print('Successfully moved')
```

```python
Listtrain=[]
Listtest=[]
for i in range(len(classes)):
    cur_class=os.listdir(os.path.join(train_dir,classes[i]))
    cur_test_class=os.listdir(os.path.join(test_dir,classes[i]))
    print(classes[i]+" "+str(len(cur_class)))
    Listtrain.append(len(cur_class))
    print(classes[i]+" "+str(len(cur_test_class)))
    Listtest.append(len(cur_test_class))
print(Listtrain)
print(Listtest)
#plt.figure(figsize=(100,1000))
#ax = fig.add_subplot(111)
fig, ax = plt.subplots()
plt.xticks(np.arange(len(classes)), classes, rotation='horizontal')

width=0.3
x = np.arange(len(classes))
ax.bar(x-width/2, Listtrain, width=0.3 ,color='b',label="train" )
ax.bar(x+width/2, Listtest, width=0.3,color='g',label="test" )
plt.legend(loc='upper left')
plt.ylabel('No of Images')
plt.xlabel('Classes')
plt.margins(0.001)
plt.title('Divided Dataset')

plt.show()
#plt.savefig("out.png")
```

```python
import keras
keras.__version__
```

```python
from keras import layers
from keras import models

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(5, activation='softmax'))
```

```python
model.summary()
```

```python
from keras import optimizers

model.compile(loss='categorical_crossentropy',
              optimizer=optimizers.RMSprop(lr=1e-4),
              metrics=['acc'])
```

```python
from keras.preprocessing.image import ImageDataGenerator

# All images will be rescaled by 1./255
train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        # This is the target directory
        train_dir,
        # All images will be resized to 150x150
        target_size=(150, 150),
        batch_size=20,
        # Since we use categorical_crossentropy loss, we need binary labels
        class_mode='categorical')
```

```python
history = model.fit_generator(
      train_generator,
      epochs=10)
```

```python
model.save('Datasets/Weights/flowers.h5')
```

```python
# Predicting se
from tensorflow.keras.models import load_model
from keras.preprocessing import image
import numpy as np

classifier = load_model('Datasets/Weights/coil-20-unproc.h5')

img=image.load_img('Datasets/flowers/test/rose/3742155164_14b557a51c_n.jpg', target_size=(150, 150,3))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
images = np.vstack([x])
prediction = classifier.predict(images, batch_size=None)
print(prediction)
print(np.argmax(prediction))
```
